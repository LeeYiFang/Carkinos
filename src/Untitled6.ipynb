```python
from pathlib import Path
import pandas as pd
import numpy as np
import sklearn
from sklearn.decomposition import PCA
import pylab
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
sanger_val_pth=Path('../').resolve().joinpath('src','sanger_cell_line_proj.npy')
sanger_val=np.load(sanger_val_pth.as_posix(),mmap_mode='r')
```

```python
sanger_val
```

```python
sanger_val=sanger_val[~np.isnan(sanger_val).any(axis=1)]
```

```python
pd.isnull(sanger_val).any(1).nonzero()[0]
```

```python
val=np.matrix(sanger_val)
```

```python
val
val.shape
```

```python
t_val=np.transpose(val)
pca= PCA(n_components=22277)
X = pca.fit_transform(t_val)
X
```

```python
propotion=[]
```

```python
t=pca.explained_variance_ratio_
sum(t[0:22277])
```

```python
propotion=[]
for n in range(0,800):
    propotion.append(sum(t[0:n]))
    print(n+1,':',sum(t[0:n]))
```
